## KNN - Euclidean distance

```python
from sklearn import datasets
import numpy as np
import pandas as pd
from math import sqrt
```

```python
# calculate the Euclidean distance between two vectors
def euclidean_distance(row1, row2):
    distance = 0.0
    # peut etre des distances a plusieur dimension
    for i in range(len(row1)-1):
        distance += (row1[i] - row2[i])**2
    return sqrt(distance)
```

```python
# Locate the most similar neighbors
def get_neighbors(train, test_row, k):
    distances = list()
    for train_row in train:
        # distance du point a tout les autres points du dataset
        dist = euclidean_distance(test_row, train_row)
        # on stock la ligne et la distance du point a tester
        distances.append((train_row, dist))
    # on les ordonnes par distance décroissante
    distances.sort(key=lambda dist: dist[1])
    neighbors = list()
    # en prenant les k premier valeurs de distance, 
    # on récuperer donc les k plus proche voisin
    for i in range(k):
        neighbors.append(distances[i][0])
    return neighbors
```

```python
iris = datasets.load_iris()
df_data = pd.DataFrame(iris.data, columns=iris.feature_names)
df_target = pd.DataFrame(iris.target, columns=["target_names"])
data = pd.concat([df_data, df_target], axis=1, sort=False)
data.head(5)
```
